# 선형 회귀 알고리즘

## 회귀(Regression)
- 현대 통계학을 이루는 큰 축
- 회귀 분석은 유전적 특성을 연구하던 영국의 통계학자 갈톤(Galton)이 수행한 연구에서 유래했다는 것이 일반론

> "부모의 키가 크더라도 자식의 키가 대를 이어 무한정 커지지 않으며, 부모의 키가 작더라도 대를 이어 자식의 키가 무한정 작아지지 않는다."

- 회귀 분석은 데이터가 많이 쌓일 수록 데이터 값이 평균과 같은 일정한 값으로 돌아가려는 경향을 이용한 통계학 기법

## 회귀의 개요
- 회귀는 여러 개의 독립변수($X$)와 한 개의 종속변수($y$) 간의 상관관계를 모델링하는 기법을 통칭
- 독립변수 $x_i$의 값에 영향을 미치는 가중치 $w_i$는 회귀 계수(Regression Coefficients)라고 한다.

$$
y = f(X) = w_1 x_1 + w_2 x_2 + \ldots + w_n x_n = \sum_{i=1}^N w_i x_i
$$

- 머신러닝에서 회귀 알고리즘/모델링은 각 피처 데이터에 부여된 회귀 계수(가중치)로 타겟값을 도출할 수 있는 수식을 세워나가는 과정이다.
- 머신러닝 회귀 예측의 핵심은 주어진 피처, 타겟 데이터 기반에서 학습을 통해 최적의 회귀 계수를 찾아내는 것이다.
  - 선형 회귀 모델의 목적은 예측값과 실제값의 RSS를 최소화하는 최적의 회귀 계수를 찾는 것

## 회귀의 유형
### 회귀 계수의 선형 여부
- 회귀 계수가 선형일 경우 : 선형 회귀(Linear Regression)
- 회귀 계수가 비선형일 경우 : 비선형 회귀
  - Tree, 딥러닝의 경우 무조건 비선형 회귀(선형 회귀의 한계를 극복함)
- 정형 데이터의 경우 대부분의 경우 비선형 회귀보다 선형 회귀의 성능이 월등히 좋다.
  - 정형 데이터 : 스키마 존재O. Type과 형식이 정해져있는 데이터. 표 형식으로 나타낼 수 있는 데이터.
  - 비정형 데이터 : 스키마 존재X. 이미지, 텍스트 데이터. 머신러닝이 아닌 딥러닝에서만 비정형 데이터를 사용한다.
  - 딥러닝은 머신러닝을 대체할 수 있는 성능을 가졌지만 무겁다는 단점이 있다. 따라서 필요에 따라 머신러닝과 딥러닝을 선택해서 사용한다.

### 독립변수의 개수
- 독립변수의 개수가 1개일 경우(피처 1개) : 단일 회귀
- 독립변수의 개수가 여러 개일 경우(피처 여러개) : 다중 회귀



- 보통 선형 다중 회귀의 경우가 많다.

## 선형 회귀의 종류
- 일반 선형 회귀(Linear Regression)
  - 예측값과 실제값의 RSS(잔차 제곱의 합)를 최소화할 수 있도록 경사 하강법을 사용하여 회귀 계수(가중치)를 최적화하는 방식
  - 규제(Regularization)를 적용하지 않은 모델
- 규제 선형 회귀
  - 릿지(Ridge) : 선형 회귀에 L2 규제를 추가한 방식
  - 라쏘(Lasso) : 선형 회귀에 L1 규제를 적용한 방식
  - 엘라스틱넷(ElasticNet) : 선형 회귀에 L2, L1 규제를 함께 결합한 모델
- 로지스틱 회귀(Logistic Regression)
  - 분류 알고리즘에 사용되는 선형 모델

## 단순 선형 회귀(Simple Regression)

$$
y_i = f(x) + Error_i = w_0 + w_1 x_i + Error_i
$$

> - $y_i$ : 실제 타겟값, $f(x)=w_0 + w_1 x_i$ : 모델이 예측한 값, $Error_i$ : 각 데이터의 오류(오차) 값, 실제값과 예측값의 차이
>
> - $w_0$ : bias. 편향, $w_1$ : 가중치, $x_i$ : 피처값

## RSS
- 각 데이터 포인트의 오류 값($Error_i$)의 제곱을 구해서 더하는 방식
- 일반적으로 미분 등의 계산을 편리하게 하기 위해서 RSS 방식으로 오류의 제곱 합을 구한다.

$$
RSS = \sum Error^2
$$

### 최적의 회귀 모델링
- 최적의 회귀 모델을 만드는 것은 전체 데이터의 잔차(오류값)의 합, 즉 RSS가 최소가 되는 모델을 만드는 것
- 동시에 RSS가 최소가 될 수 있는 최적의 회귀 계수 $w$를 찾는 것
- RSS는 회귀식의 중심 변수는 독립변수 $X$, 종속변수 $Y$가 아닌 회귀 계수 $w$임을 인지해야 한다.
  - 머신러닝 모델에 학습 데이터로 입력되는 $X$, $Y$는 이미 알고 있는 데이터이기 때문에 RSS에서 모두 상수로 간주한다.
  - RSS가 최소가 될 수 있는 최적의 미지수 $w$를 찾는 것이 목표

### 회귀의 비용 함수(Cost Function)
- 회귀에서 RSS는 비용(Cost)
- $w$ 변수(회귀 계수)로 구성되는 RSS를 비용 함수라고 한다.(RSS가 비용 함수 중 하나)
- 비용 함수는 손실 함수(Loss Function)이라고도 한다.
- 머신러닝 회귀 알고리즘은 데이터를 계속 학습하면서 이 비용 함수가 반환하는 값인 오류값(Loss)을 지속해서 감소
- 최종적으로 더 이상 감소하지 않는 최소의 오류값을 구하는 것이 학습의 목적
$$
RSS(w_0, w_1) = \frac{1}{N}\sum_{i=1}^N (y_i - (w_0 + w_1 x_i))^2
$$

> - $w_o, w_1$ : 회귀 계수, RSS의 변수
>
> - $N$ : 학습 데이터 건수




## 경사 하강법(Gradient Descent) - 비용, 손실 최소화하는 방법
- 회귀 알고리즘의 목적인 Loss(Cost)를 지속적으로 감소하기 위해 사용하는 방법
- $w$ 파라미터 개수가 적다면 고차원 방정식(선형 연립 방정식)으로 비용(손실) 함수가 최소가 되는  $w$ 변숫값을 의사역행렬 방식으로 구할 수 있다.
  - 하지만 $w$ 파라미터 개수가 많다면(피처 데이터가 너무 많다면) 각 데이터에 대한 의사역행렬을 구하는데 컴퓨터의 물리적인 파워 등에서 한계가 존재한다.
- 경사 하강법은 이러한 고차원 방정식에 대한 문제를 해결해주면서 비용(손실)함수 RSS를 최소화하는 방법을 직관적으로 제공하는 뛰어난 방식이다.
  - 즉, 경사 하강법은 RSS 최적화 방법 중 하나
  - 최적화(Optimization) : 해를 바로 구하는 것이 아니라, 해에 가까운 계수를 점진적으로 찾아가는 방법 및 과정

- Gradient Descent : 점진적 하강
  - 경사 하강법은 점진적으로 반복적인 계산을 통해 $w$ 파라미터 값을 업데이터 하면서 오류 값이 최소가 되는 $w$ 파라미터를 구하는 방식이다.

$$
w = w + η(- \frac{\partial L}{\partial w}) = w - η \frac{\partial L}{\partial w}
$$

> - $w$ : 가중치, η : learning_rate(학습률), $\partial L$ : 손실 함수 RSS의 미분값, $\partial w$ : 가중치 $w$의 미분값

- 초기 학습으로 구한 $w$의 기울기의 수식을 구한다.
- $\frac{\partial L}{\partial w}$는 $w$의 기울기 진행 방향을 확인할 수 있다.
$$
y = \frac{\partial L}{\partial w} w
$$
- $w$를 갱신하는 방향인  $w$의 기울기의 반대방향($- \frac{\partial L}{\partial w}$)이다.
- 학습률 η가 미분값을 조정해주는 역할을 한다.
  - η값이 너무 작으면 $w$가 제대로 갱신되지 않을 수 있다.
  - η값이 너무 크면 미분값이 발산하여 RSS가 최소인 최적의 $w$를 구하기 어렵다.
  - 보통 η 초기값으로 0.001 또는 0.01 정도로 설정한다.

- 경사 하강법은 반복적으로 비용(손실) 함수의 반환값인 예측값과 실제 값의 차이 Loss($Error_i$)가 작아지는 방향성을 가지고 $w$ 파라미터를 지속적으로 갱신해 나간다.
- 오류값이 더 이상 작아지지 않으면, 그 오류 값을 최소 비용으로 판단하고 그 때의 $w$ 파라미터를 최적의 파라미터로 반환한다.
  - 미분된 함수의 기울기 $\frac{\partial L}{\partial w}$가 더이상 감소하지 않는 지점



## `LinearRegression` 클래스
```python
sklearn.linear_model.LinearRegression(fit_intercept=True, normalize=False)
```

- 예측값과 실제값의 RSS를 최소화하는 OLS(Ordinary Least Squares) 추정 방식으로 구현한 클래스
  - OLS : RSS를 최소화하는 방식 중 하나
  - 즉 `LinearRegression` 클래스는 OLS를 구현한 회귀 알고리즘 클래스

- `fit_intercept` : 절편 값을 계산할 것인지 여부 결정. False로 설정 시 intercept가 0으로 지정된다.
  - 보통 `fit_intercept=True`를 사용
- `normalize` : `fit_intercept=True`일 경우 회귀를 수행하기 전 입력 피처 데이터 세트를 표준화(Standard Scaling)함
  - `fit_intercept=False`인 경우에는 무시된다.
  - 대체로 사용하지 않고 `normalize=False`로 해둔다. 선형 회귀 모델에서 데이터 전처리 과정에서 표준화, 정규화 작업을 필수적으로 선행하기 때문이다.

```python
## 훈련된 LinearRegression 모델의 회귀 계수 확인
LinearRegression().coef_

## 훈련된 LinearRegression 모델의 절편 확인
LinearRegression().intercept_
```
- 훈련(fit)시에 회귀 계수(Coefficeints)인 $W$를 `coef_` 속성에 저장한다.
- 절편(bias, intercept)는 `intercept_`속성에 저장한다.

## 선형 회귀의 다중 공선성(mulit-collinearity) 문제
- 일반적으로 선형 회귀는 입력 피처 데이터의 독립성에 많은 영향을 받는다.
  - 피처 데이터가 선형 종속일 때 문제 발생
- 다중 공선성 : 피처 간의 상관 관계가 매우 높은 경우 분산이 매우 커져 오류에 매우 민감해져 예측이 제되로 되지 않는 현상
  - 분산이 커지면 두 피처 데이터 간의 진행 방향이 매우 비슷하게 되어 오류에 민감해짐
  - ex. 건물의 평수&제곱 미터, 자동차의 무게&연비
- 일반적으로 상관관계가 높은 피처 데이터가 많은 경우
  - 독립적인 중요한 피처만 남기고 제거
  - 또는 규제를 적용한다.
- 따라서 데이터 간의 상관성을 확인하는 작업이 반드시 필요하다.


## 회귀 평가 지표

### 1. MAE(Mean Absolute Error)
- 실제값과 예측값의 차이를 절대값으로 변환해 평균
- 피처의 아웃라이어(이상치)이 많을 때 유리

$$
MAE = \frac{1}{N}\sum_{i=1}^N |y_i - \hat y_i|
$$

### 2. MSE(Mean Squared Error) ⭐
- RSS를 의미. 머신러닝에서는 MSE라고 한다.
- 실제값과 예측값의 차이를 제곱해 평균
- 오차 제곱하는 과정에서 오차를 를 더 과장해서 보여주기 위해  MAE보다 오차를 수정하기 쉽다.

$$
MSE = \frac{1}{N}\sum_{i=1}^N (y_i - \hat y_i)^2
$$

### 3. MSLE(Mean Squared Log Error)
- MSE에 Log를 적용한 값
- 결정값이 클 수록 오류값이 커지기 때문에 일부 큰 오류값들로 인해 전체 오류값이 커지는 것을 막아준다.
- 타겟에 아웃라이어(이상치)가 많은 경우 사용
- 혹은 빈도수나 타겟값의 차이가 너무 심할 때 사용
  - ex. 타겟값이 각각 10, 1000으로 나올 때 Log를 적용하면 1, 3이 됨

$$
MSLE = \frac{1}{N}\sum_{i=1}^N (log(y_i+1) - log(\hat y_i+1))^2
$$

> - $log0=-\infty$ 인 경우를 방지하기 위해 log 함수에서 1을 더한다.

### 4. RMSE(Root Mean Sqaured Error)⭐
- MSE는 오류의 제곱을 구하여 실제 오류 평균보다 더 커지는 특성이 있기 떄문에 MSE에 루트를 씌운 것
- MAE, MSE에 비해 큰 오류에 상대적인 패널티를 부여한다.
  - 여러 오류값 중 다른 값에 비해 큰 오류값이 존재할 경우 RMSE는 전반적으로 MAE보다 높다. 때문에 MAE보다 RSS를 갱신하기 쉽다.
- MSE는 머신러닝에서 오류 계산을 위해 사용하고 RMSE는 오류에 대해 확인용, 설명용으로 사용한다.
- 보통 MSE를 만들고 MSE에 루트를 씌운 RMSE를 같이 만들어 준다.

$$
RMSE = \sqrt{\frac{1}{N}\sum_{i=1}^N (y_i - \hat y_i)^2}
$$

### 5. RMSLE(Root Mean Squared Log Erro)
- RMSE에 Log를 적용한 것
- 결정값이 클 수록 오류값이 커지기 때문에 일부 큰 오류값들로 인해 전체 오류값이 커지는 것을 막아준다.

$$
MSLE = \sqrt{\frac{1}{N}\sum_{i=1}^N (log(y_i+1) - log(\hat y_i+1))^2}
$$

## 6. $R^2$
- 분산(평균으로 부터 떨어진 정도) 기반으로 **예측 성능을 평가**한다.
- 실제 값의 분산 대비 예측값의 분산 비율을 지표로 한다.
- 회귀에서 **정확도**를 나타내는 개념이다.
- 1에 가까울 수록 예측 정확도가 높아지게 된다.

$$
R^2 = \frac{예측값 variance}{실제 값 variance}
$$


## 회귀 평가 API
- 평가 방법에 대한 사이킷런 API 및 `cross_val_score`, `GridSearchCV`에서 평가시 사용되는 `scoring` 파라미터의 적용 값

|평가지표|사이킷런 평가지표 API|Scoring 함수 적용값|
|---|---|---|
|MAE|`metrics.mean_absolute_error`|`neg_mean_absolute_error`|
|MSE|`metrics.mean_squared_error`|`neg_mean_squred_error`|
|MSLE|`metrics.mean_squred_log_error`|`neg_mean_squared_log_error`|
|RMSE|`metrics.mean_squared_error` 사용하고 `squared=False`로 파라미터 지정|`neg_root_mean_squred_error`|
|$R^2$|`metrics.r2_score`|`r2`|

### `Scoring` 함수에 회귀 평가 지표 적용시 유의 사항
- MSE의 사이킷런 `scoring` 파라미터 값은 `neg_mean_squred_error`를 적용해 음수값을 반환
  - Negative(음수) 값을 가진다는 의미
  - MAE의 절대값의 합이기 때문에 음수가 될 수 없다.
- 회귀 알고리즘에서 Score는 Loss(오차값)에 대한 지표이기 때문에 Score가 낮을 수록 좋은 지표이다. 사이킷런의 `Scoring` 함수가 Score 값이 클 수록 좋은 평가 결과로 자동 평가하기 때문에 평가 지표 값에 -1를 곱해 Negative를 만들어 작은 오류 값이 더 큰 숫자로 인식되게 한다.
- `metrics.mean_squared_error()`와 같은 사이킷런 평가 지표 API는 정상적인 양수의 값을 반환한다. 하지만 `Scoring` 함수의 `scoring` 파라미터 값은 `neg_mean_squred_error`가 의미하는 것은 `-1 * metrics.mean_squared_error()`이다. 이 점을 유의하여 코드를 작성해야 한다.

## 단일 선형 회귀 모델 실습
- 보스턴 주택 가격 예측

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')  # 사이킷런 1.2 부터는 보스턴 주택가격 데이터가 없어진다는 warning 메시지 출력 제거

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

boston_df = pd.DataFrame(
    data=data,
    columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
)

boston_df['PRICE']=target
boston_df
````

### Feature와 Feature의 상관관계

- Feature간의 상관관계 확인하는 이유 : 어떤 Feature가 Target을 예측하는데 있어서 결정적인 역할을 하는지 확인하기 위해서
- Feature와 Feature의 상관성이 낮을 수록 서로 독립적인 Feature이기 때문에 머신러닝에 투입하기 좋다고 해석할 수 있다.

````python
## 타겟(PRICE)와 Fetuare의 상관관계를 절대값으로 확인 => 0 ~ 1 사이의 값을 반환
np.abs(
    boston_df.corr()["PRICE"]
).sort_values()
````

- `LSTAT` 피처가 타겟인 `PRICE`와 제일 강한 상관관계를 가지고 있다는 것을 확인

````python
## 타겟(PRICE)와 Fetuare의 상관관계를 절대값 없이 다시 확인 => -1 ~ 1 사이의 값을 반환
boston_df.corr()["PRICE"].sort_values()
````

````python
# 상관 계수의 절대값을 확인해야 0 ~ 1사이의 값으로 반환하여 확인하기 더 용이하다.
corr_matrix_pos = np.abs(boston_df.corr())

## 히트맵으로 전체 데이터의 상관성 추세를 시각화해서 확인
plt.figure(figsize=(10, 10))

sns.heatmap(
    data=corr_matrix_pos,
    annot=True
)

plt.show()
````

- 히트맵에서 1에 가까운 밝은 부분 위주로 확인하면 된다.
  - 주대각방향은 데이터 자신에 대한 상관관계를 나타내는 것으로 1이 나온다.  제외하고 본다.
- 0.91로 나온 부분이 제일 높은 상관관계로 보인다.
  - `RAD` 피처, `TAX` 피처는 서로 상관성이 매우 높다는 것을 확인
  - 모델의 성능이 떨어져 상관관계가 높은 피처 중 하나를 탈락시킬 경우, 타겟과의 상관관계가 비교적 낮은 피처를 포기하는 것이 좋다.
    - `RAD` 피처 보다는 `TAX` 피처를 탈락시키는 것이 좋을 수 있다.(0.38 < 0.47)
- 추후 비즈니스 전문가의 조언으로 피처에 대한 상관관계를 파악하는 단계도 있다.

````python
## 데이터 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    boston_df.drop('PRICE', axis=1),
    boston_df['PRICE'],
    test_size=0.2,
    random_state=42
)
````

### 하이퍼 파라미터를 지정하지 않을 경우

````python
## 선형 회귀 모델 객체 생성
from sklearn.linear_model import LinearRegression
lr_reg = LinearRegression()

## 모델 훈련
lr_reg.fit(X_train, y_train)
````

````python
## 모델 예측
y_pred = lr_reg.predict(X_test)
y_pred[:3]
````

````python
## 모델 예측 성능 평가하기
from sklearn.metrics import mean_squared_error, r2_score

# MSE와 RMSE 확인
# accuracy_score와 비슷한 형식으로 확인 가능
mse = mean_squared_error(y_test, y_pred) # RSS
rmse = np.sqrt(mse) # RMSE - MSE에 루트값

print(f'MSE : {mse}')
print(f'RMSE : {rmse}')
````

````python
# R2 확인 - 회귀에서의 정확도에 해당
r2 = r2_score(y_test, y_pred)

print(f'R2 : {r2}')
````

- $R^2$는 예측의 분산값이 실제 데이터들의 분산을 얼마나 반영을 하는가를 나타내는 지표로 회귀에서 정확도에 해당하는 개념이다.
- `r2 = 1.0`이면 완벽하게 실제 데이터를 예측 했다는 의미
- 현재 모델의 예측 정확도는 약 67%로 볼 수 있다.

### `cross_val_score`로 교차 검증을 할 경우

````python
## 교차 검증
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    LinearRegression(),
    X_train,
    y_train,
    cv=5,
    scoring='neg_mean_squared_error'
)

scores
````

- `scoring='neg_mean_squared_error'` score 점수에 대한 지표를 Loss로 보기 떄문에 음수로 나타내어 score가 낮을 수록 좋은 score가 나오게 한다.

````python
# 위에 나온 score는 음수이기 때문에, 원래대로 score를 되돌리기 위해 -1를 곱해준다.
# 보통 score는 높을 수록 좋은 점수여야 하기 때문에
mse_scores = -1 * scores
rmse_scores = np.sqrt(mse_scores)

# 평균 rmse
rmse_mean = np.mean(rmse_scores)
rmse_mean
````

- 모델의 오차가 약 5라고 볼 수 있다.
