## 벡터의 기하학적 의미

### 기하학
점, 직선, 곡면, 면, 부피 등 **공간에 있는 성질(치수, 모양, 상대적 위치 등)에 관한 것**
- 다루는 대상 : 점, 선, 면, 도형, 공간

### $N$차원 벡터 $a$의 기하학적 의미
- $N$차원 벡터 $a$는 **$N$차원 공간**에서 점 또는 원점과 점까지 연결한 화살표

#### 1. 벡터 $a$의 값으로 표시되는 점(point)
- 아래 벡터 $a$는 2차원 공간에서 $x$좌표가 $a_1$, $y$좌표가 $a_2$인 $(a_1, a_2)$ 점으로 생각할 수 있다.

$$
a = \begin{bmatrix} a_1 \\ a_2 \end{bmatrix}
$$


#### 2. 원점과 벡터 $a$의 값으로 표시되는 점을 연결한 화살표(arrow)
- 벡터를 화살표로 생각하는 경우, **길이와 방향을 고정시킨 채 평행이동(translation)할 수 있다.**
- **벡터 $a$를 평행이동한 벡터는 $a$와 같은 벡터라고 본다.**
  - 단지 화살표의 시작점이 원점이 아닌 다른 점부터 시작한 것이다. 화살표 연결의 시작점인 원점은 단순히 기준점의 역할이다.

- 아래 맷플롯립으로 시각화한 그래프에서 빨간색 점이 벡터 $a$를 2차원 공간에서 점으로 표현한 것이고, 화살표로 표시된 부분이 화살표로 표현한 것이다.

```python
import numpy as np
import matplotlib.pylab as plt

plt.rc("font", size=18)  # 그림의 폰트 크기를 18로 고정
gray = {"facecolor": "gray"}
black = {"facecolor": "black"}
red = {"facecolor": "red"}
green = {"facecolor": "green"}
blue = {"facecolor": "blue"}

a = np.array([1, 2])
plt.plot(0, 0, 'kP', ms=20)
plt.plot(a[0], a[1], 'ro', ms=20)
plt.annotate('', xy=[-0.6, 1.6], xytext=(0.2, 0.7), arrowprops=gray)
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=black)
plt.annotate('', xy=a + [-1, 1], xytext=(-1, 1), arrowprops=black)
plt.text(0.35, 1.15, "$a$")
plt.text(1.15, 2.25, "$(1,2)$")
plt.text(-0.7, 2.1, "$a$")
plt.text(-0.9, 0.6, "translation")
plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-2.4, 3.4)
plt.ylim(-0.8, 3.4)
plt.show()
```

## 벡터의 길이

- 벡터 $a$의 길이는 벡터 $a$의 놈($\|a\|$)으로 정의

$$
\|a\| = \sqrt{a^T a} = \sqrt{a1^2 + \ldots + a_N^2}
$$

- 넘파이에서 `linalg` 서브 패키지의 `norm` 함수로 벡터의 길이를 계산할 수 있다.

```python
a = np.array([1, 2])
b = np.array([[1], [2], [3]])

np.linalg.norm(a), np.linalg.norm(b)
```

## 스칼라와 벡터의 곱

- **스칼라의 부호**($+, -$)는 **벡터의 방향성**을 표현한다.
    - $+$ : 벡터의 방향이 변하지 않음
    - $-$ : 벡터의 반대 방향이 됨
- 스칼라값에 해당하는 **실수의 절대값**만큼 **벡터의 길이**가 변화한다.

- 아래 맷플롯립으로 시각화한 그래프에서 스칼라 $2$를 곱한 벡터 $2 a$는 빨간색 화살표, 스칼라값 $-1$을 곱한 벡터 $-1 a$는 파란색 화살표로 표시

```python
a = np.array([1, 2])
b = 2 * a
c = -a
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=red)
plt.text(0.8, 3.1, "$2a$")
plt.text(2.2, 3.8, "$(2, 4)$")
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)
plt.text(0.1, 1.3, "$a$")
plt.text(1.1, 1.4, "$(1, 2)$")
plt.plot(c[0], c[1], 'ro', ms=10)
plt.annotate('', xy=c, xytext=(0, 0), arrowprops=blue)
plt.text(-1.3, -0.8, "$-a$")
plt.text(-3, -2.5, "$(-1, -2)$")
plt.plot(0, 0, 'kP', ms=20)
plt.xticks(np.arange(-5, 6))
plt.yticks(np.arange(-5, 6))
plt.xlim(-4.4, 5.4)
plt.ylim(-3.2, 5.2)
plt.show()
```

## 단위벡터(unit vector)

- **길이가 1인 벡터**
 
$$
a = \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \;\;\;\; b = \begin{bmatrix} 0 \\ 1 \end{bmatrix}, \;\;\;\; c = \begin{bmatrix} \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} \end{bmatrix}
$$

```python
a = np.array([[1], [0]])
b = np.array([[0], [1]])
c = np.array([[1 / 2**0.5], [1 / 2**0.5]])

np.linalg.norm(a), np.linalg.norm(b), np.linalg.norm(c)
```

#### 벡터의 정규화
- 벡터를 단위 벡터로 만드는 과정
    - 화살표의 길이에 상관없이 **방향만 고려하기 위해 사용**
- $0$벡터가 아닌 임의의 벡터 $x$를 정규화하면 벡터 $\frac{x}{\|x\|}$이다.
    - 벡터 $\frac{x}{\|x\|}$는 벡터 $x$와 **같은 방향을 가리키는 단위벡터**를 의미한다.
    - 분모로 오는 놈이 $0$이 되지 않기 위해 0벡터가 아니어야 한다.

$$
\frac{x}{\|x\|} = \frac{x}{\sqrt{x^Tx}}
$$

## 벡터의 합

- 두 벡터의 합은 벡터이다.
- 두 벡터의 합은 원점에서 벡터 $a$ 만큼 이동한 후, 벡터 $b$만큼 이동한 위치이다.
- 두 벡터의 합은 벡터 $b$를 더하고자 하는 벡터 $a$의 끝점으로 평행이동했을 때 이동한 벡터가 가리키는 점의 위치이다.
    - 둘 중 어느 벡터를 평행이동해도 결과는 같다. 

```python
a = np.array([1, 2])
b = np.array([2, 1])
c = a + b
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=gray)
plt.annotate('', xy=c, xytext=a, arrowprops=gray)
plt.annotate('', xy=c, xytext=(0, 0), arrowprops=black)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.plot(c[0], c[1], 'ro', ms=10)
plt.text(0.35, 1.15, "$a$")
plt.text(1.15, 0.25, "$b$")
plt.text(0.1, 2.6, "$b$(translation)")
plt.text(1.25, 1.45, "$c$")
plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.show()
```

### 벡터의 선형조합
- 선형조합 : 여러 개의 벡터에 스칼라값을 곱한 후 모두 더한 것

$$
c_1 x_1 + c_2 x_2 + \ldots + c_N x_N
$$

> - $c_i$는 스칼라 계수, 벡터 $x$

- **선형조합을 통해 벡터를 원하는 방향 및 길이로 변화시킬 수 있다.**
    - 기하학관점에서 아래 예시의 선형조합 $0.5x_1 + x_2$은 벡터 $x_1$의 파워가 $0.5$만큼 줄어든다고 한다.

$$
c = \begin{bmatrix} c_1 & c_2 \end{bmatrix} = \begin{bmatrix} 0.5 & 1 \end{bmatrix}, \;\;\;\; x_1 = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \;\;\;\; x_2 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}
$$

$$
c_1 x_1 + c_2 x_2 = 0.5 x_1 + x_2 
$$

## 벡터의 차

- 벡터의 차 $a - b = c$는 벡터 $b$에서 벡터 $a$로 이동한 벡터
    - 벡터 $b$에 벡터 $a - b$를 더하면 벡터 $a$가 되어야 하기 때문

$$
a - b = c
$$

$$
a = b + c
$$

- 위 식을 다시 정리하면 벡터 $a$는 벡터 $b$에서 벡터 $c$만큼 이동한 벡터로 볼 수 있다.

- 벡터 $a-b$의 길이는 벡터 $a$와 벡터 $b$ 사이의 거리가 된다.

-  $a$와 $b$ 사이의 거리($a-b$의 길이)를 측정하는 방식
    1. 맨하튼 거리 : 벡터가 좌표평면에서 몇 칸씩 움직이는지 확인하는 방식
    2. 유클리드 거리

### 유클리드 거리
- 벡터의 차의 놈을 거리로 보는 방식

$$
\|c\| = \|a-b\|
$$

## 벡터의 내적과 삼각함수

- 두 벡터의 내적의 결과값($a^Tb$)인 스칼라값은 두 가지 방식으로 구할 수 있다.
    1. 직접 내적하여 구하기
    2. 두 벡터 사이의 각도값(기울기)인 $\theta$를 이용해 구하기

$$
a^Tb = \|a\| \|b\| \cos\theta
$$

> - $\|a\|$ : 벡터 a의 길이, $\|b\|$ : 벡터 b의 길이

$$
\cos\theta = \frac{a^Tb}{\|a\| \|b\|}
$$

## 정규 직교

- $N$개의 단위벡터($v_1,v_2,\ldots,v_N$) 모두가 서로 직교하는 것

$$
\|v_i\| = 1 \leftrightarrow v_i^Tv_i = 1
$$

- $i$번째 벡터의 길이는 $1$이기 때문에 벡터의 제곱합은 $1$이된다.

$$
v_i^Tv_j = 0
$$

- 다른 벡터와의 내적은 항상 $0$이 된다.

## 코사인 유사도

- 코사인 유사도가 높을수록 두 벡터가 비슷하다고 본다.
- 코사인 유사도가 낮을수록 두 벡터가 비슷하지 않다고 본다.
- 코사인 유사도의 최소값 $0$일 때 두 벡터는 전혀 다른 벡터로 본다.
- 코사인 유사도의 최대값 $1$일 때 두 벡터는 같은 방향을 보고있는 같은 벡터로 본다.

### 코사인 거리
- 두 벡터의 코사인 거리가 0에 가까워질 수록 두 벡터가 비슷하다고 본다.

## 벡터의 분해와 성분

- $a+b=c$일 때 $a$와 $b$는 $c$의 성분이라고 한다.
- $c$를 $c$의 성분으로 나타내는 것을 분해라고 한다.

- $c$의 성분인 $a$와 $b$는 무수히 많은 해를 가진다.

## 투명 성분과 직교성분
$$
a = a^{\Vert b}  + a^{\perp b} 
$$

- 벡터 $a$는 $a$의 투명성분($a^{\Vert b}$)과 직교성분($a^{\perp b} $)의 합으로 분해된다.

## 투명 성분

- 벡터 $a$와 다른 벡터 $b$와 평행하는 벡터 $a$의 성분

## 직교 성분

- 벡터 $a$와 다른 벡터 $b$와 직교하는 벡터 $a$의 성분
